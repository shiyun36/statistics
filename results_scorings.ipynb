# Traitement des grilles d'évaluation

```python
def extract_notes_from_tables(doc_path):
    doc = docx.Document(doc_path)

    notes_per_table = []

    for table in doc.tables:
        notes_row = table.rows[1]
        notes = []
        for cell in notes_row.cells:
            try:
                note = float(cell.text.strip())
                notes.append(note)
            except ValueError:
                notes.append(None)

        notes_per_table.append(notes)

    return notes_per_table
```

```python
doc_path = '/content/Grille de scoring with new dem cog M8O7A5.docx'

notes = extract_notes_from_tables(doc_path)

print(len(notes))

for i in range(len(notes)):
    notes[i].pop(0)

df = pd.DataFrame(notes)
df.index = [f"Table {i+1}" for i in range(len(notes))]
df.columns = [f"Critère {i+1}" for i in range(df.shape[1])]

csv_path = 'extracted-notes-new-dem-M8O7A5.csv'
df.to_csv(csv_path, index=True)

print(f"Les notes ont été enregistrées dans {csv_path}. Voici les données extraites :")
print(df)
```
